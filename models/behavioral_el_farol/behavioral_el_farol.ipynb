# Behavioral El Farol Model
This notebook demonstrates a structured behavioral version of the El Farol Bar problem using a modular decision pipeline.

```python
from core.model import BehavioralElFarolModel
import matplotlib.pyplot as plt
```

```python
N_AGENTS = 100
CAPACITY = 60
STEPS = 50
SEED = 42
```

```python
model = BehavioralElFarolModel(
    n_agents=N_AGENTS,
    capacity=CAPACITY,
    seed=SEED
)

attendance = []

for _ in range(STEPS):
    model.step()
    attendance.append(model.current_attendance)
```

```python
plt.figure(figsize=(6, 4))

plt.plot(attendance, label="Attendance")
plt.axhline(y=CAPACITY, linestyle="--", label="Capacity")

plt.xlabel("Step")
plt.ylabel("Attendance")
plt.title("Behavioral El Farol Model")

plt.legend()
plt.show()
```

```python
# Observations:
# - Initial steps may show extreme fluctuations due to lack of information
# - Over time, attendance stabilizes around capacity
# - The system exhibits bounded oscillation rather than perfect coordination
#
# This behavior emerges from:
# - heterogeneous prediction strategies
# - probabilistic decision-making
```
